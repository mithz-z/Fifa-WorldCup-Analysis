```python
def posic(mergedf):
    winner = mergedf['winner'].unique()
    winner.sort()
    sec = mergedf['second'].unique()
    sec.sort()
    third = mergedf['third'].unique()
    third.sort()
    fourth = mergedf['fourth'].unique()
    fourth.sort()
    return winner, sec, third, fourth
```

```python
def posi(mergedf, c):
    winner = mergedf[mergedf['winner'] == c]
    sec = mergedf[mergedf['second'] == c]
    third = mergedf[mergedf['third'] == c]
    fourth = mergedf[mergedf['fourth'] == c]
    
    return winner, sec, third, fourth
```
